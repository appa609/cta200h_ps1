```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x*(x-1)

a = -1
b = 1
n = 10000

x = np.linspace(a,b,n)
y = f(x)
plt.plot(x,y)

dx = float(b-a)/n

plt.show()
```

```python
def differencequotient(g,x,dt):
    return (g(x+dt)-g(x))/dt

def diff(g,x,e):
    #print("entered")

    D1 = -9999
    D2 = -9999
    d1 = -9999
    d2 = -9999
    
    for i in range(0,15):
        #print(i)
        dt = 10**-i
        D1 = differencequotient(g,x,dt)
        D2 = differencequotient(g,x-dt,dt)
        
        if abs(d1-D1) > 10**-e:
            d1 = D1
            #print("iterate D1: " + str(D1))
        
        if abs(d2-D2) > 10**-e:
            d2 = D2
            #print("iterate D2: " +str(D2))
            
        if abs(d1-D1) < 10**-e and abs(d2-D2) < 10**-e and abs(D1-D2)<10**-e:
            return (D1+D2)/2
    print("Derivative did not converge at "+str(x))    
    return (D1+D2/2)
    
            

            
```

```python
print(differencequotient(f,1,dx))
```

```python
dy = np.empty(n)
for i in range(n):
    dy[i] = diff(f,x[i],5)
plt.plot(x,dy)
plt.show()
```

```python
def sinx(x):
    return x**2*np.sin(1/x)
    
    
y = sinx(x)
plt.plot(x,y)
plt.show()
```

```python
dy = np.empty(n)
for i in range(n):
    dy[i] = diff(sinx,x[i],6)
plt.plot(x,dy)
plt.show()
```
